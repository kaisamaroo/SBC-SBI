This notebook is used to generate and store the Hadamard matrices used in this section

```python
from scipy.linalg import hadamard
import numpy as np
```

```python
results_path = "/Users/Lieve/Documents/Masters Project/SBC-SBI/results/toy_examples/unif_norm_high_d_data/hadamard_matrices" 
```

```python
# We will create rectangular hadamard matrices of shape k by k
k_list = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384] # Up to 2**14
```

```python
for k in k_list:
    k_by_k_hadamard = hadamard(k) / np.sqrt(k) # Create k by k Hadamard matrix
    save_path = results_path + f"/square_{k}.npy"
    np.save(save_path, k_by_k_hadamard)
```
