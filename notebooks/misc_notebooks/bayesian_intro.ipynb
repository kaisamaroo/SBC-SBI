```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
np.random.seed(1)

# Define prior:
a, b = 5,20
prior_pdf = lambda p: beta.pdf(p, a, b) # Beta(alpha, alpha) prior

n = 50

# Posterior
def posterior_pdf(p, s):
    # s = number of successes
    # n-s = number of failures
    post_alpha = a + s
    post_beta = b + n - s
    return beta.pdf(p, post_alpha, post_beta)



# Plot
fig, ax = plt.subplots(figsize=(10,4))
p = np.linspace(0, 1, 500)
ax.plot(p, prior_pdf(p), color="black", label="Prior")

s1 = 10
ax.plot(p, (lambda p: posterior_pdf(p, s1))(p), color="red", label=f"Posterior given that we observe \n {s1} males and {n-s1} females.")

s2 = 25
ax.plot(p, (lambda p: posterior_pdf(p, s2))(p), color="green", label=f"Posterior given that we observe \n {s2} males and {n-s2} females.")

s3 = 45
ax.plot(p, (lambda p: posterior_pdf(p, s3))(p), color="blue", label=f"Posterior given that we observe \n {s3} males and {n-s3} females.")

ax.set_xlabel(r"$p$ (probability of a given customer being male)")
ax.set_ylabel("Probability density")
ax.set_title(r"Prior and various posterior distributions over $p$")
ax.legend()
plt.show()
```
